```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("_DataVisualization/_amazon/amazon.csv", encoding="ISO-8859-1")

df.columns
```

```python
state = df.groupby("state").mean().drop("year", axis=1)

plt.barh(state.index, state.number)
```

```python
year = df.groupby("year").sum()

plt.xticks(year.index, rotation="vertical")
plt.plot(year.index, year.number)
```

```python
year_diff = year.diff(periods=1, axis=0)

plt.xticks(year_diff.index, rotation="vertical")
plt.bar(year_diff.index, year_diff.number)
```

```python
month_brasil = ["Janeiro",  "Fevereiro", "Maro",     "Abril",
                "Maio",     "Junho",     "Julho",     "Agosto",
                "Setembro", "Outubro",   "Novembro",  "Dezembro"]
month_brasil = pd.DataFrame(month_brasil, columns=["month"])

month = df.groupby("month").mean().drop("year", axis=1).reset_index(level=0, inplace=False)
month = pd.merge(month_brasil, month)

plt.xticks(rotation="vertical")
plt.plot(month.month, month.number)
```
